```python
import pandas as pd
import numpy as np
import keras.models
from keras.models import Sequential
from keras.layers import Dense
from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
```

```python
data=load_breast_cancer()
```

```python
data.keys()
```

```python
print(data['DESCR'])
```

```python
print(data['data'])
```

```python
print(data['target_names'])
print(data['filename'])
```

```python
pd.DataFrame(data['data'],columns=data['feature_names'])
```

```python
feature=data['data']
label=data['target']
```

```python
scaling=StandardScaler()
feature=scaling.fit_transform(feature)
```

```python
pd.DataFrame(feature,columns=data['feature_names'])
```

```python
feature.shape
```

```python
X_train=feature[:500]
X_test=feature[500:535]
X_val=feature[535:]
y_train=label[:500]
y_test=label[500:535]
y_val=label[535:]
```

```python
model= Sequential()
model.add(Dense(15,activation='relu',input_dim=30))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy']) 
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=5,validation_split=.1)
```

```python
model.evaluate(X_test,y_test)
```

```python
for i in range(10):
  test=np.reshape(X_val[i],(1,30))
  if(float(model.predict(test)[0][0])>0.5):
    print('-bening')
  else:
    print('-malignant')
  if(y_val[i]>0.5):
    print('*Bening')
  else:
    print('*Malignant')
  print('--------------')
```
